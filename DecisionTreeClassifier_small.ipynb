```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
```

```python
# Load the dataset
file_path = 'sat_train.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Split the dataset into features (X) and target (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=365)
```

```python
# Create a decision tree classifier with pruning
pruning_hyperparameter = 0.02
dt_classifier = DecisionTreeClassifier(random_state=365, ccp_alpha=pruning_hyperparameter)

# Fit the model to the training data
dt_classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier, filled=True)
plt.show()
```

```python
# Predict the target variable for the test set
y_pred = dt_classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
